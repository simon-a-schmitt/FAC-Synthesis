```python
from datasets import load_dataset
```

```python

# use name="sample-10BT" to use the 10BT sample
fw = load_dataset("HuggingFaceFW/fineweb-edu", name="CC-MAIN-2024-10", split="train", streaming=True)
```

```python
fw = load_dataset("HuggingFaceFW/fineweb-edu", name="sample-10BT", split="train", streaming=True).shuffle(seed=42, buffer_size=10000).take(200000)
```

```python
for example in fw.take(5):
    print(example["text"])
```

```python
from tqdm import tqdm

output_file = "fineweb_edu_sample.txt"

with open(output_file, "w", encoding="utf-8") as f:
    for example in tqdm(fw, total=200000, desc="Writing documents"):
        text = example["text"]
        # Escape newlines so each document is on one physical line
        text_escaped = text.replace("\n", "\\n")
        f.write(text_escaped + "\n")
```
